```python
import os
import optuna
import pandas as pd
from matplotlib import pyplot as plt
```

```python
study_summaries = optuna.study.get_all_study_summaries(storage="sqlite:///db.sqlite3")

studies = {summary.study_name: summary for summary in study_summaries}
print(studies.keys())

# plot several figures in a grid
def plot_grid(figures, nrows = 1, ncols=1):
    fig, axeslist = plt.subplots(ncols=ncols, nrows=nrows)
    for ind,title in zip(range(len(figures)), figures):
        axeslist.ravel()[ind].imshow(figures[title], cmap=plt.gray())
        axeslist.ravel()[ind].set_title(title)
        axeslist.ravel()[ind].set_axis_off()
    plt.tight_layout() # optional
    plt.show()

plt.figure(figsize=(10,10))

# create a directory if it doesn't exist
def create_dir(dir_name):
    dir = f"/visualizations/{dir_name.replace(':','_').replace(' ','_')}/"
    if not os.path.exists(dir):
        os.makedirs(dir)

def create_study_visualization(study_name):
    print(study_name)
    overfit_threshold = 0.90
    df = optuna.study.load_study(study_name, storage="sqlite:///db.sqlite3").trials_dataframe().dropna()

    if df.shape[0] < 3:
        return

    df.rename(columns = {'values_0': 'training_error', 'values_1': 'validation_error'}, inplace = True)
    df.duration = df.duration.dt.total_seconds()

    # overfitting is defined as the difference between the training and validation mse being greater than 0.1 (10%) and the validation mse being greater than 0.9 (90%)
    def is_overfit(df) -> bool:
        return df['training_error'] < 0.05 and df['validation_error'] /  df['training_error'] > 0.97

    # underfitting is defined as the difference between the training and validation accuracy being less than 0.1 (10%) and or the validation accuracy being less than 0.9 (90%)
    def is_underfit(df) -> bool:
        return df['validation_error'] > overfit_threshold

    print(study_name)
    df = df[df['state'] == 'COMPLETE']
    df['is_overfit'] = df.apply(is_overfit, axis=1)
    df['is_underfit'] = df.apply(is_underfit, axis=1)
    df['overfit_degree'] = df.apply(lambda row: row['validation_error']/row['training_error'], axis=1)
    df['is_optimal'] = df.apply(lambda trial: not is_overfit(trial) and not is_underfit(trial), axis=1)

    ax = df[df.state == 'COMPLETE'].corr()['validation_error'].plot(kind='bar', title='Validation Error Correlation')
    ax.set_xlabel("Hyperparameter")
    ax.set_ylabel("Correlation with Validation MSE")

    cols = df.columns
    cols = cols.drop(['number', 'datetime_start', 'datetime_complete', 'state'])
    feature_cols = cols.drop(['validation_error', 'training_error', 'is_underfit', 'is_optimal', 'is_overfit'])
    target_col = ['validation_error', 'training_error']

    def example_plot(ax, fontsize=12, i=0):
        x, y = i // 2, i % 2
        ax.scatter(df[feature_cols[x]], df[target_col[y]])

        ax.locator_params(nbins=3)
        ax.set_xlabel(feature_cols[x], fontsize=fontsize)
        ax.set_ylabel(target_col[y], fontsize=fontsize)
        ax.set_title(f"{target_col[y]} vs. {feature_cols[x]}", fontsize=fontsize)


    fig, axs = plt.subplots(len(target_col), len(feature_cols), layout="constrained", figsize=(15,10))
    for i, ax in enumerate(axs.flat):
        example_plot(ax, i=i)
    fig.savefig(f"plots/{study_name.replace(':','_').replace(' ','_')}")
    display(df[df['is_overfit'] == True])

for study_name in studies.keys():
    create_study_visualization(study_name)
```

```python
best_models = []
study_names = []
study_values = []
for study_name in studies.keys():
    df = optuna.study.load_study(study_name, storage="sqlite:///db.sqlite3").trials_dataframe().dropna()
    df.rename(columns = {'values_0': 'training_error', 'values_1': 'validation_error'}, inplace = True)
    study_names.append(study_name)
    study_values.append(df['validation_error'].min())

best_models = pd.DataFrame({'study name': study_names, 'validation error': study_values}).set_index('study name').sort_values('validation error')
ax = best_models.plot(kind='bar', title='KDDCUP99 Best Model')
ax.set_xlabel("study name")
ax.set_ylabel("Validation MSE")
plt.savefig('plots/best_model.png')
plt.show()

best_models
```
